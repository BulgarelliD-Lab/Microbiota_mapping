* Problem - data we have for introgression lines have alleles as A, T etc. We need A, B (ref alternative). So get this information

* Convert lines to "A" and "B" genotypes. A is reference (Barke), "B" is wild allele.

```python
genotype_9k_subset.head()
```

```python
name = "Int_52"
line_of_interest = QTL2[name]
Barke = QTL2["Barke"]
bases = {"A","T","C","G"}
converted = []
for i in line_of_interest.index:
    #print(f"{Int_17[i]}, {Barke[i]}")
    if line_of_interest[i] == Barke[i]:
        converted.append("A")
    elif line_of_interest[i] in bases:
        converted.append("B")
    elif "/" in str(line_of_interest[i]) and str(Barke[i]) in str(line_of_interest[i]):#If het and ref allele in call
        converted.append("H")
    else:
        converted.append("-")

line_of_interest_c = pd.Series(converted,index=line_of_interest.index)       
```

```python
similarity_matrix = pd.DataFrame({"A":[1,0.5,0], "H": [0.5,1,0.5], "B": [0, 0.5, 1]}, index=["A", "H", "B"])
```

* First make a similarity matrix to refer to. E.g an A allele compared to an A allele is the same (1)

```python
similarity_matrix
```

* Now go though each of the mapping lines and calculate a score out of 1 for similarity against Int_17

```python
similarities = {}
possible_alleles = {"A","B","H"}
for i in range(2,51):
    similarity = 0
    total = 0
    line = i
    for n, allele in enumerate(line_of_interest_c):
        #print(f"Int 17 allele is {allele}")
        line_allele = QTL2.at[n,line]
        #print(f"For marker {marker}, allele_AB is {allele_AB} and line allele is {line_allele}")
        #print(f"line allele is {line_allele}")
        if line_allele not in possible_alleles or allele not in possible_alleles:
            #print(f"Error! line allele is |{line_allele}|")
            continue
        similarity += similarity_matrix.at[line_allele, allele]
        total += 1
    try:
        frequency = similarity/total
    except:
        frequency = 0
    similarities.setdefault(name, []).append(frequency)

similarities = pd.DataFrame(similarities, index=[i for i in range(2,51)])
        
        

        
    
```
