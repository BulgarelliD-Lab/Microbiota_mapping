* First import previous table from R:

```python
import pandas as pd
import os

R_table = "/cluster/db/mecoulter/RNAseq2/RNAquant_analysis_result/candidate_gene_complete_info.csv"
variant_table = "/cluster/db/mecoulter/RNAseq2/high_impact_variant_annotation.csv"

table = pd.read_csv(R_table)
table = table.rename(columns={"gene": "Genes"})
table.head()



```

```python
table2 = pd.read_csv(variant_table)

table2.head()
```

```python
merged_table = pd.merge(table,table2,how="left", on="Genes")
len(merged_table)
merged_table.head()
```

```python
len(merged_table)
merged_table.columns
```

```python
merged_table = merged_table.drop(["Unnamed: 0_x","Unnamed: 0_y"],axis=1)
merged_table.head()
```

```python
merged_table = merged_table.rename(columns={"raw_p": "Raw p", "adj_p": "Adj. p", "function.": "Predicted function", "GO.IDs":"GO ids"})
merged_table.head()
```

```python
#Write to file
merged_table = merged_table.drop(["GO ids"],axis=1)#Drop this for now, too many values
merged_table.to_csv(path_or_buf="/cluster/db/mecoulter/RNAseq2/RNAseq_variant_candidates.csv",header=True)


```
